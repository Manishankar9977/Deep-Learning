```python
import pandas as pd
df = pd.read_csv('IRIS.csv')
df
```

```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import OneHotEncoder
```

```python
iris_data = load_iris()
iris_data.target_names
```

```python
x = iris_data.data
y_ = iris_data.target.reshape(-1, 1)
```

```python
encoder = OneHotEncoder(sparse=False)
y = encoder.fit_transform(y_)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.20)
```

```python
model = Sequential() 
model.add(Dense(50,input_dim=4,activation='relu')) 
model.add(Dense(25, activation='relu')) 
model.add(Dense(12, activation='relu')) 
model.add(Dense(3, activation='softmax')) 

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

print('Neural Network Model Summary: ')
print(model.summary())

model.fit(train_x, train_y, epochs=200, batch_size=10) 
results = model.evaluate(test_x, test_y) 

print('Final test set loss: {:4f}'.format(results[0]))
print('Final test set accuracy: {:4f}'.format(results[1]))
```

```python
#now using the data set 
x1 = df.drop('species', axis = 1)
y1= df['species'] 
```

```python
yy=y1.values
ryy=yy.reshape(-1,1)
encoder = OneHotEncoder(sparse=False)
y11 = encoder.fit_transform(ryy)
```

```python
model = Sequential() 
model.add(Dense(100,input_dim=4,activation='relu')) 
model.add(Dense(50, activation='relu')) 
model.add(Dense(25, activation='relu')) 
model.add(Dense(3, activation='softmax')) 

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(x1, y11, epochs=200, batch_size=10) 
_, accuracy = model.evaluate(x1,y11) 


print('Accuracy: %.2f' % (accuracy*100))
```
